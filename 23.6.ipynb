```python
import folium
from geopy.geocoders import Nominatim
import osmnx as ox
import networkx as nx
import geopandas as gpd
import branca
m=folium.Map([10.756123413786124, 106.66376464090362], zoomstart=13)
folium.Marker([10.756123413786124, 106.66376464090362], popup='Trường THPT Hùng Vương', tooltip='Click here').add_to(m)
G=ox.graph_from_point((10.756123413786124, 106.66376464090362),dist=1000,network_type='drive')
stats = ox.basic_stats(G)
html = f"""
    <h1>Thông tin khu vực</h1>
    <p><b>Địa điểm:</b> Trường THPT Hùng Vương</p>
        <b>Số nút giao: {stats['n']}
        <b>Số đoạn đường:</b> {stats['m']}
        <b>Chiều dài TB:</b> {stats['street_length_avg']:.2f} m
"""
print(html)
iframe = branca.element.IFrame(html=html, width=250, height=180)
popup = folium.Popup(iframe, max_width=300)
nodes, edges = ox.graph_to_gdfs(G)
folium.GeoJson(edges).add_to(m)
m

```
