```python
import folium
from geopy.distance import geodesic
kho = [10.7561278,106.6616627]
noi_den = [[10.766251246999259, 106.67675465485806], [10.753438471838338, 106.67903111754373], [10.75059629362271, 106.6509551024173],[10.768161458078165, 106.65593480104616]]
duong_di = [kho] + sorted(noi_den, key=lambda x: geodesic(kho, x).meters)
m = folium.Map(location=kho, zoom_start=15)
folium.Marker(kho, icon=folium.Icon(color='red')).add_to(m)
for p in noi_den: folium.Marker(p, icon=folium.Icon(color='blue')).add_to(m)
folium.PolyLine(duong_di, color='red', weight=8).add_to(m)
m
```
