```python
#bài tập 23.1
import folium
m = folium.Map(location=[10.7562744,106.6638299], zoom_start=13)
school = folium.FeatureGroup('Shool')
hospital = folium.FeatureGroup('Hospital')
mall = folium.FeatureGroup('mall')
folium.Marker([10.7562744,106.6638299], popup='THPT Hùng Vương - Trường công lập uy tín tại TP.HCM', tooltip='More detail').add_to(school)
folium.Marker([10.7565468,106.665875], popup='BV Truyền Máu Huyết Học<br>Chuyên khoa huyết học hàng đầu TP.HCM', tooltip='More detail').add_to(hospital)
folium.Marker([10.755264,106.6629435], popup='Đại học Y Dược TP.HCM<br>Trường đào tạo y khoa hàng đầu Việt Nam', tooltip='More detail').add_to(school)
folium.Marker([10.7561278,106.6616627], popup='Bệnh viện Hùng Vương<br>Bệnh viện phụ sản uy tín tại TP.HCM',tooltip='More detail').add_to(hospital)
folium.Marker([10.756114,106.6627253], popup='Hùng Vương Plaza<br>Trung tâm thương mại sầm uất tại TP.HCM',tooltip='More detail').add_to(mall)
folium.Marker([10.7556384,106.6646794], popup='BV Đại học Y Dược TP.HCM<br>Bệnh viện đa khoa chất lượng cao',tooltip='More detail').add_to(hospital)
school.add_to(m)
hospital.add_to(m)
mall.add_to(m)
folium.LayerControl().add_to(m)
m
```
