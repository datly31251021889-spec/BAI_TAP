```python
#bài tập 23.2
import folium
from geopy.distance import geodesic
from geopy.geocoders import Nominatim
dia_diem=["cho Ben Thanh, Ho Chi Minh City",
"Dinh doc lap, Ho Chi Minh City",
"Nha tho Duc ba, Ho Chi Minh City",
"Cong vien Tao Dan, Ho Chi Minh City",
"Saigon Centre, Ho Chi Minh City",
"Pho di bo Nguyen Hue, Ho Chi Minh City",
"Benh vien Cho Ray, Ho Chi Minh City",
"Dai hoc y duoc thanh pho Ho Chi Minh, Ho Chi Minh City",
"san bay quoc te Tan Son Nhat, Ho Chi Minh City",
"Landmark 81, Ho Chi Minh City"]
m=folium.Map(location=[10.761327233494937, 106.66828289481502], zoom_start=13)
geolocator = Nominatim(user_agent="geoapi", timeout=10)
folium.Marker([10.761327233494937, 106.66828289481502], popup='Đại học Kinh tế TP.HCM', tooltip='More detail').add_to(m)
for i in dia_diem:
  point_a= (10.761327233494937, 106.66828289481502)
  point_b=geolocator.geocode(i)
  point_b = (point_b.latitude, point_b.longitude)
  folium.Marker(point_b, popup=f'{i}', tooltip='More detail', icon=folium.Icon(color='red')).add_to(m)
  distance=geodesic(point_a,point_b).km
  folium.PolyLine([point_a, point_b],popup=f'cách UEH cơ sở B {distance}km').add_to(m)
m
```
