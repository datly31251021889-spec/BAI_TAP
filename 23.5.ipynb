```python
#Bài tập 23.5
import folium
nha_kho = {"Kho A": [10.7565468,106.665875], "Kho B": [10.773389071422123, 106.70117186824456], "Kho C": [10.815804844208422, 106.66420072883588],
}
m = folium.Map(location=[10.7565468,106.665875], zoom_start=13)
folium.Marker(nha_kho['Kho A'], popup='BV Truyền Máu Huyết Học', icon=folium.Icon(color="green")).add_to(m)
folium.Circle(nha_kho['Kho A'], radius=3000, color="blue", fill=True, fill_opacity=0.1).add_to(m)
folium.Marker(nha_kho['Kho B'], popup='SaiGon center', icon=folium.Icon(color="green")).add_to(m)
folium.Circle(nha_kho['Kho B'], radius=5000, color="blue", fill=True, fill_opacity=0.1).add_to(m)
folium.Marker(nha_kho['Kho C'], popup='Sân bay Tân Sơn Nhất', icon=folium.Icon(color="green")).add_to(m)
folium.Circle(nha_kho['Kho C'], radius=10000, color="blue", fill=True, fill_opacity=0.1).add_to(m)
m
```
